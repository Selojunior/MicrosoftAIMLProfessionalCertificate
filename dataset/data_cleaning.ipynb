```python
import pandas as pd 
import numpy as np
```

```python

df1 = pd.read_csv("Season_2025-2026_fixed.csv")
```

```python
df1.head()
```

```python
df2 = pd.read_csv("All_Players_1992-2025_fixed.csv")
```

```python
df2.head()
```

```python
def clean_columns(df: pd.DataFrame) -> pd.DataFrame:
    df.columns = (
        df.columns
        .astype(str)
        .str.replace('"', '', regex=False)  # remove "
        .str.strip()                        # remove spaces around names
    )
    return df

df1 = clean_columns(df1)
df2 = clean_columns(df2)

df1.columns[:10], df2.columns[:10]
```

```python
print(len(df1.columns), len(df2.columns))
print(df1.columns.equals(df2.columns))
```

```python
df_final = pd.concat([df2, df1], ignore_index=True) 

df_final.shape
df_final.head()
```

```python
df_final.tail()
```

```python
df_final.to_csv("all_players_1992-2026.csv", index=False)
```
